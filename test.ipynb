```python
from tqdm import tqdm
import copy
import math
def hits_articleSort(terms):
    naive=naive_articleSort(terms)
    hadict={}
    artlist=naive
    lennaive=len(naive)
    size=min(100,lennaive)
    maxtimes=200
    minerror=0.0001
    for i in range((size)):
        #print('naive['+str(i)+']='+naive[i])
        artlist+=simartdict[naive[i]]
    artlist=list(set(artlist))
    #print('artlist:',len(artlist),artlist)
    for i in artlist:
        hadict[i]=1
    #hits
    error=1000
    for i in range(maxtimes):
        #if(i<=4):
            #print('This is the {}th time'.format(i))
            #print(hadict)
        if(error<minerror):
            break
        tmp=copy.deepcopy(hadict)
        for a in artlist:
            norm=0
            score=0
            for j in simartdict[a]:
                if j in tmp.keys() and j is not a:
                    score+=tmp[j]
            #print('score=',score,end=' ')
            hadict[a]=score
            norm+=pow(score,2)
        norm=math.sqrt(norm)
        #print('norm=',norm)
        for a in artlist:
            hadict[a]/=norm
            error=abs(tmp[a]-hadict[a])
    hits=sorted(naive,key=lambda ind:hadict[ind],reverse=True)
    print('--------------------hits--------------------------\n',hits,hadict)
    return hits
```

```python
def naive_articleSort(terms):
    print('Now enter naive_articleSort()')
    respond=[]
    
    articlenum=len(data.body)
    score=[0 for i in range(articlenum)]
    for keyword in tqdm(terms):
        if keyword not in wordlist:
            continue
        index=wordlist.index(keyword)
        for i in range(articlenum):
            score[i]+=eval(data['tf-idf'][i])[index]
    score_index=list(zip(score,range(len(score))))
    score_index.sort(key=lambda t:t[0],reverse=True)
    for i in range(max_return):
        if score_index[i][0]==0:
            break
        ind=score_index[i][1]
        respond.append(ind)
        #return respond
    #return [('haha','adhjajcha'),('Oscar','Great!')]
    print(len(respond),respond)
    return respond
```
